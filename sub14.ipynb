```python
import csv
head = {
         1:'ID',
         2:'Attrition',
         3:'YearsAtCompany'
    }

file = open('trial1.csv', 'w', newline ='',encoding="utf-8")
with file: 
        
        # identifying header 
        header = [ 'ID', 'Attrition', 'YearsAtCompany']
        
        writer = csv.DictWriter(file, fieldnames = header) 
        # writing data row-wise into the csv file 
        writer.writeheader() 
        
        ran=[]
        answer=[]
        for i in range(3):
            ran.append(0)
            
        for l in range(0,240):
            ran[0] = ids[l]
            ran[1] = proans1[l]
            ran[2] = y_predtest2[l]
            answer.append(list(ran))
            for i in range(3):
                ran.append(0)
        
        for i in range(0,240):
            class my_dictionary(dict):
                
                # __init__ function  
                def __init__(self):  
                    self = dict()  
                # Function to add key:value  
                def add(self, key, value):  
                    self[key] = value    
            dict_obj = my_dictionary()  
            
            for j in range(0,3):
                dict_obj.add(head[j+1], answer[i][j]) 
                
            
            #print(dict_obj)
            writer.writerow(dict_obj)
```
